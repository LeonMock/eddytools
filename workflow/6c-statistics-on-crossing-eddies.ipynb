```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# parameters
minimum_duration = 5
first_or_last_crossing = 'first'

experiment_name = 'INALT20r.L120-KRS006'#'INALT60.L120-KRS0020'
data_resolution = '1d'

tracking_start = "2012-01-01"
tracking_end = "2012-12-31"
```

```python
# detection parameters
OW_thr_factor =-0.3
if experiment_name.startswith("INALT60"):
    Npix_min, Npix_max = 720, 18000
    sigma = 15
    wx = 600
elif experiment_name.startswith("INALT20"):
    Npix_min, Npix_max = 80, 2000
    sigma = 5
    wx = 200
```

```python
depth= 0  #corresponding to... 
depth_index = 0 # tracking only on surface
```

```python
outpath = f'/gxfs_work/geomar/smomw523/eddytools/results/{experiment_name}/smoothed/{sigma}/{data_resolution}/depth-{depth}/'   ## !! SMOOTHED !!
```

```python
df = pd.read_csv(outpath + f'Good-hope-{first_or_last_crossing}-crossings_min-{minimum_duration}d_{tracking_start.replace("-", "")}_{tracking_end.replace("-", "")}_OW{np.abs(OW_thr_factor)}_Npix-{Npix_min}-{Npix_max}_rolling-{wx}.csv')
```

```python
df
```

```python
crossings=df
crossings[(crossings['section'] == 'Ansorge-GHS') & (crossings['scale'] >= 50) & (crossings['duration'] >= 50)]
```

```python
hist_columns = ['duration', 'amp', 'scale', 'area']
titles = ['duration [days]', 'vorticity amplitude', 'average radius [km] ', 'area [km]']
type_colors = {'cyclonic': 'blue', 'anticyclonic': 'red'}
sections = ['Ansorge-GHS']#, 'Schmidt-GHS']

fig, axes = plt.subplots(1, 4, figsize=(18, 4))  
axes = axes.flatten()

for j, section in enumerate(sections):
    df_section = df[df['section'] == section]
    
    for i, col in enumerate(hist_columns):
        bins = np.histogram_bin_edges(df[col], bins='auto')
        bins = np.linspace(np.round(bins[0] / (10 ** np.round(np.log10(bins[0]))), 1) * 10 ** np.round(np.log10(bins[0])),
                           np.round(bins[-1] / (10 ** np.round(np.log10(bins[-1]))), 1) * 10 ** np.round(np.log10(bins[-1])), 11)
        
        for type_label, color in type_colors.items():
            sns.histplot(df_section[df_section['type'] == type_label][col], bins=bins, kde=True, line_kws={'linewidth': 3},
                         label=type_label, ax=axes[i + j * 4], color=color, alpha=0.4)
        
        #axes[i].set_title(f'{col}',fontsize=20)
        axes[i+j*4].set_xlabel(f'{titles[i]}',fontsize=15)
        axes[i+j*4].set_ylabel('')
        #axes[i+j*4].legend()

    print(f'{section}, existed at start: {np.round(100*df_section["exist_at_start"].sum()/len(df_section),1)}%')
    print(f'{section}, not terminated at end: {100-np.round(100*df_section["terminated"].sum()/len(df_section),1)}%')

plt.tight_layout()
plt.show()
```
