```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
```

```python
mesh_mask = xr.open_dataset('/gxfs_work/geomar/smomw523/inalt60/data/INALT60.L120-KRS0020/OW-prep/2_INALT60.L120-KRS0020_mesh_mask.nc') 
mesh_mask_46 = xr.open_dataset('/gxfs_work/geomar/smomw044/prep_data_iAtlantic_LMock/INALT20.L46-KFS119/1_INALT20.L46-KFS119_mesh_mask_iMirabilis.nc')
indices = np.concatenate((range(0, 11, 10),range(18, 25, 6),range(29, 34, 4),range(36, 40, 3),range(41, 120, 2)))
depth_information = [(round(mesh_mask.nav_lev.values[i]), i) for i in indices]
print(len(depth_information),depth_information)
```

```python
for depth in [tupel[0] for tupel in depth_information]:
    plt.hlines(depth,0,np.max(np.diff(mesh_mask_46.nav_lev)),color='gray',linewidth=.5)
plt.gca().invert_yaxis()
plt.plot(np.diff([tupel[0] for tupel in depth_information]),[tupel[0] for tupel in depth_information][1:])
plt.plot(np.diff(mesh_mask.nav_lev),mesh_mask.nav_lev[1:],linestyle='dotted')
plt.plot(np.diff(mesh_mask_46.nav_lev),mesh_mask_46.nav_lev[1:],linestyle='dotted')
```
